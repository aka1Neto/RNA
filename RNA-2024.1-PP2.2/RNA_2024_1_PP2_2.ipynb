### Função de ativação sigmoide

$$f(x) = \frac{1}{1 + e^{-\beta \cdot x}}$$

- Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada
- Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função sigmoide
- Em um mesmo gráfico, ilustre 3 valores distintos para $\beta$:
  - $\beta = 0.7$ na cor laranja
  - $\beta = 0.5$ na cor verde
  - $\beta = 0.3$ na cor vermelha
  - Use legendas no gráfico

```python
# Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada

def sigmoid(teta, x):
  return 1 / (1 + np.e ** (-teta * x))
```

```python
# Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
# Plote o gráfico da função no intervalo de entrada previamente especificado

plt.plot(dados, sigmoid(1, dados), linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid")
plt.show()
```

```python
# Implemente uma função que retorne o valor da derivada da função sigmoide

def derivative_sig(teta, x):
  sigmoid = 1 / (1 + np.e ** (-teta * x))
  return sigmoid * (1 - sigmoid)

plt.plot(dados, derivative_sig(1, dados), linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Derivative Sigmoid")
plt.show()
```

```python
# Em um mesmo gráfico, ilustre 3 valores distintos para  β :

plt.plot(dados, sigmoid(0.7, dados), linestyle='-', color='orange', label='β = 0.7')
plt.plot(dados, sigmoid(0.5, dados), linestyle='-', color='green', label='β = 0.5')
plt.plot(dados, sigmoid(0.3, dados), linestyle='-', color='red', label='β = 0.3')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoide")
plt.show()
```

### Função de Ativação ReLU (Unidade Linear Retificada)

$$f(x) = max(0,x)$$

- Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função ReLU

```python
# Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada

def relu(x):
  return np.maximum(0, x)
```

```python
# Plote o gráfico da função no intervalo de entrada previamente especificado

plt.plot(dados, relu(dados), linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("ReLU")
plt.show()
```

```python
# Implemente uma função que retorne o valor da derivada da função ReLU

def derivative_relu(x):
  return np.where(x >= 0, 1, 0)

plt.plot(dados, derivative_relu(dados), linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Derivative ReLU")
plt.show()
```

### Função de Ativação Gaussiana

$$f(x) = \exp \left( \frac{-(x - \mu)^2}{2\sigma^2} \right)$$

- $\mu$ é o ponto médio (centro) da curva
- $\sigma$ denota o raio de abertura
- Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão


Considere também os seguintes valores ao plotar a função

- Parâmetros nomeados $\mu = 0$ e $\sigma = 1$
- $\mu = 0, \sigma = 0.3$
- $\mu = 0, \sigma = 0.5$
- $\mu = 0, \sigma = 0.7$
- $\mu = 0, \sigma = 0.9$
- $\mu = 0, \sigma = 1.1$

```python
# Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada

def gauss(x, sigma, mu=0):
  return np.exp((-(x - mu) ** 2) / (2 * sigma ** 2))
```

```python
# Plote o gráfico da função no intervalo de entrada previamente especificado

plt.plot(dados, gauss(dados, 0.3), linestyle='-', color='red', label='σ = 0.3')
plt.plot(dados, gauss(dados, 0.5), linestyle='-', color='yellow', label='σ = 0.5')
plt.plot(dados, gauss(dados, 0.7), linestyle='-', color='green', label='σ = 0.7')
plt.plot(dados, gauss(dados, 0.9), linestyle='-', color='blue', label='σ = 0.9')
plt.plot(dados, gauss(dados, 1), linestyle='-', color='purple', label='σ = 1')
plt.plot(dados, gauss(dados, 1.1), linestyle='-', color='black', label='σ = 1.1')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gauss")
plt.show()
```

```python
# Implemente uma função que retorne o valor da derivada da função em questão

def derivative_gauss(x, sigma, mu=0):
  return -((x - mu) / sigma) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

plt.plot(dados, derivative_gauss(dados, 0.3), linestyle='-', color='red', label='σ = 0.3')
plt.plot(dados, derivative_gauss(dados, 0.5), linestyle='-', color='yellow', label='σ = 0.5')
plt.plot(dados, derivative_gauss(dados, 0.7), linestyle='-', color='green', label='σ = 0.7')
plt.plot(dados, derivative_gauss(dados, 0.9), linestyle='-', color='blue', label='σ = 0.9')
plt.plot(dados, derivative_gauss(dados, 1), linestyle='-', color='purple', label='σ = 1')
plt.plot(dados, derivative_gauss(dados, 1.1), linestyle='-', color='black', label='σ = 1.1')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Derivative Gauss")
plt.show()
```

### Função de Ativação Leaky ReLU

$$f(x) = \left\{\begin{array}{cc} 0.01\cdot x & x < 0\\
x & x \geq 0
\end{array}
\right.$$

- Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função Leaky ReLU (consulte a literatura!)

```python
# Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada

def leaky_relu(x):
  return np.where(x < 0, 0.01 * x, x)
```

```python
# Plote o gráfico da função no intervalo de entrada previamente especificado

plt.plot(dados, leaky_relu(dados), linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Leaky ReLU")
plt.show()
```

```python
# Implemente uma função que retorne o valor da derivada da função Leaky ReLU

def derivative_leaky_relu(x):
  return np.where(x < 0, 0.01, 1)

plt.plot(dados, derivative_leaky_relu(dados), linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Derivative Leaky ReLU")
plt.show()
```
